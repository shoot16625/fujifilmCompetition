# yolo学習用のデータを作成
## ラベルありの画像だけ抽出し，imagesフォルダとlabelsフォルダを作成
## ローカルで実行したあと，作成されたフォルダをzipファイルにしてgoogle driveのdataフォルダに配置
## 実行環境：ローカルのpython3.6にrequirements.txtをpipでインストール

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
import os
import re
%matplotlib inline
```

## ローカルのinputフォルダ内にコンペデータを配置

```python
path = "input/train_anns/"
train_ann_filenames = os.listdir(path)
train_ann_filenames.sort()
```

```python
# ラベルのあるデータのみ抽出してimagesフォルダに保存
path = "input/train_images/"
n = 0
for target in train_ann_filenames:
    target = re.sub("json", "jpg", target)
    I = cv2.imread(path+target)
    cv2.imwrite("images/"+target, I)
    n += 1
print(n)
```

```python
char_dic = {"'": 10,
            '9': 9,
            '3': 3,
            '7': 7,
            '2': 2,
            '0': 0,
            '6': 6,
            '1': 1,
            '8': 8,
            '5': 5,
            '4': 4,
            '/': 11,
            '.': 12,
            'yy': 13,
            'mm': 14,
            'dd': 15}
```

```python
# YOLO学習用のデータを作成して，labelsフォルダに保存
path = "input/train_anns/"
n = 0
for target in train_ann_filenames:
    I = cv2.imread("input/train_images/"+re.sub("json", "jpg", target))
    h, w, c = I.shape
    df_label = pd.read_json(path+target)
    texts = []
    for bboxes in df_label["bboxes"]:
        locs = [value for value in bboxes.values()]
#         YOLO形式変換
        x_center = (locs[3] + locs[1])/2
        x_center = x_center / w
        y_center = (locs[4] + locs[2]) / 2
        y_center = y_center / h
        x_width = (locs[3] - locs[1])/w
        y_width = (locs[4] - locs[2]) / h
        texts.append(
            list(map(str, [char_dic[locs[0]], x_center, y_center, x_width, y_width])))
    target = re.sub("json", "txt", target)
    with open("labels/"+target, mode='w') as f:
        f.writelines('\n'.join([" ".join(line) for line in texts]))
    n += 1
print(n)
```
