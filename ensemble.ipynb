```
import re
import pandas as pd
```

```
df1 = pd.read_csv("result_yolo_20000_0.23_new.csv", header=None)
```

```
df2 = pd.read_csv("result_cnn_colab.csv",
                  header=None)
```

```
# yoloとcnnの出力結果の差
n = 0
for i in range(len(df1)):
    if type(df1.at[i, 1]) != float or type(df2.at[i, 1]) != float:
        if df1.at[i, 1] != df2.at[i, 1]:
            print(df1.at[i, 1], "/", df2.at[i, 1])
            n += 1
print(n)
```

```
# yoloの結果をcnnの結果を用いて補正していく
for i in range(len(df1)):
    #     yoloがnanでCNNが文字列
    if type(df1.at[i, 1]) == float and type(df2.at[i, 1]) != float:
        print("変更：", df1.at[i, 1], "/", df2.at[i, 1])
        df1.at[i, 1] = df2.at[i, 1]
#     yoloが文字列でCNNがnan
    elif type(df1.at[i, 1]) != float and type(df2.at[i, 1]) == float:
        if len(set(df1.at[i, 1])) == 1:
            print("変更：", df1.at[i, 1], "/", df2.at[i, 1])
            df1.at[i, 1] = df2.at[i, 1]
#     yoloが文字列でCNNが文字列
    elif type(df1.at[i, 1]) != float and type(df2.at[i, 1]) != float:
        tmp1 = df1.at[i, 1].split(" ")
        tmp2 = df2.at[i, 1].split(" ")
#         CNNを信頼
        if len(tmp1) != 3 and len(tmp2) == 3:
            print("変更：", df1.at[i, 1], "/", df2.at[i, 1])
            df1.at[i, 1] = df2.at[i, 1]
        elif len(tmp1) == 3 and len(tmp2) == 3:
            #             文字列の入れ替えが起こっているときは無視
            if tmp1[0] == tmp2[0] and tmp1[1] == tmp2[2] and tmp1[2] == tmp2[1]:
                print("文字いれかえ：", df1.at[i, 1], "/", df2.at[i, 1])
            elif tmp1[0] == tmp2[1] and tmp1[1] == tmp2[0] and tmp1[2] == tmp2[2]:
                print("文字いれかえ：", df1.at[i, 1], "/", df2.at[i, 1])
            elif "".join(tmp1)[:3] == "".join(tmp2)[:3] and len(tmp2[0]) < 3 and len(tmp2[1]) < 3 and df1.at[i, 1] != df2.at[i, 1]:
                if tmp1[2] != tmp2[2]:
                    print("くっつき変更：", df1.at[i, 1], "/", df2.at[i, 1])
                    df1.at[i, 1] = df2.at[i, 1]
            else:
                if df1.at[i, 1] != df2.at[i, 1]:
                    print("変更：", df1.at[i, 1], "/", df2.at[i, 1])
                    df1.at[i, 1] = df2.at[i, 1]
```

```
# 両方Nanでないデータを抽出
filled_indexes = []
for i in df1.index:
    if type(df1.at[i, 1]) != float or type(df2.at[i, 1]) != float:
        filled_indexes.append(i)
```

```
# 前後のデータを用いて，結果を補正
n = 0
data_pre1 = ""
data_pre2 = ""
for index, i in enumerate(filled_indexes):
    new_data = []
    data_now1 = df1.at[i, 1]
    data_now2 = df2.at[i, 1]
    try:
        data_next1 = df1.at[filled_indexes[index+1], 1]
        data_next2 = df2.at[filled_indexes[index+1], 1]
    except:
        data_next1 = " "
        data_next2 = " "
#    前後のデータが合致しているとき，中央のデータも合わせる
    if len(data_now1.split(" ")) == 3 and len(data_pre1.split(" ")) == 3 and len(data_next1.split(" ")) == 3 and type(data_now2) == float:
        for j in range(3):
            if data_pre1.split(" ")[j] == data_next1.split(" ")[j] and data_now1.split(" ")[j] in data_pre1.split(" ")[j]:
                new_data.append(data_pre1.split(" ")[j])
            else:
                new_data.append(data_now1.split(" ")[j])
        predict = " ".join(new_data)
#     前後のデータが合致しているとき，中央のサイズが2
    elif len(data_now1.split(" ")) == 2 and len(data_pre1.split(" ")) == 3 and len(data_next1.split(" ")) == 3 and type(data_now2) == float:
        for j in range(3):
            if data_pre1.split(" ")[j] == data_next1.split(" ")[j]:
                new_data.append(data_pre1.split(" ")[j])
            else:
                new_data.append("X")
#         うまいこと空白をうめていく
        if new_data == ["X", "X", "X"]:
            new_data = data_now1.split(" ")
        elif re.search(data_now1+" X", " ".join(new_data)):
            del new_data[2]
        elif new_data[0] == data_now1.split(" ")[0] and new_data[1] == "X":
            new_data[1] = data_now1.split(" ")[1]
        elif new_data[1] == data_now1.split(" ")[1] and new_data[0] == "X":
            new_data[0] = data_now1.split(" ")[0]
        elif data_now1.split(" ")[0] in new_data[1] and new_data[2] == "X":
            new_data[2] = data_now1.split(" ")[1]
        elif new_data[1] == "X" and new_data[2] == "X" and new_data[0] not in data_now1.split(" "):
            if len(data_now1.split(" ")[0]) == 4:
                new_data[1] = data_now1.split(" ")[1]
            else:
                new_data[1:] = data_now1.split(" ")
        elif new_data[0] == "X" and new_data[1] == "X" and new_data[2] not in data_now1.split(" "):
            new_data[:2] = data_now1.split(" ")
        elif new_data[0] == "X" and new_data[2] == "X" and new_data[1] not in data_now1.split(" "):
            new_data[0] = data_now1.split(" ")[0]
            new_data[2] = data_now1.split(" ")[1]

        try:
            for k in range(3):
                new_data.remove("X")
        except:
            pass
        predict = " ".join(new_data)
    else:
        predict = data_now1

    if predict != data_now1:
        print(data_now1, "/", predict, "trial:", index)
        df1.at[i, 1] = predict
    data_pre1 = df1.at[i, 1]
    data_pre2 = df2.at[i, 1]
    n += 1
```

```
# 同一ユーザの予測結果を利用して，補正
data_info = "test_"
file_ids = list(set(re.findall("id[0-9]+", " ".join(df1[0]))))
file_ids.sort()
for file_id in file_ids:
    f = re.findall(data_info+file_id +
                   "_no[0-9]+\.jpg", " ".join(df1[0]))

#     文字列パターンを数え上げる
    df1_person_predicts = [list(df1[df1[0] == n][1])[0] for n in f]
    df1_number_count_dic = {}
    for i, data in enumerate(df1_person_predicts):
        if type(data) != float:
            tmp = data.split(" ")
            if len(tmp) == 3:
                try:
                    df1_number_count_dic[tmp[0]+" "+tmp[1]] += 1
                except:
                    df1_number_count_dic[tmp[0]+" "+tmp[1]] = 1
                try:
                    df1_number_count_dic[tmp[1]+" "+tmp[2]] += 1
                except:
                    df1_number_count_dic[tmp[1]+" "+tmp[2]] = 1
                try:
                    df1_number_count_dic[data] += 1
                except:
                    df1_number_count_dic[data] = 1

    for i, data in enumerate(df1_person_predicts):
        if type(data) != float:
            tmp = data.split(" ")
            if len(tmp) == 2:
                if not re.search("9[0-9]", data):
                    x, y, c = 0, 0, 0
                    for j in range(10):
                        try:
                            x = df1_number_count_dic["9"+str(j)+" "+tmp[0]]
                            if x - y > 0:
                                y = x
                                c = "9"+str(j)
                        except:
                            pass
                    for j in range(9):
                        for k in range(2):
                            try:
                                x = df1_number_count_dic["20" +
                                                         str(k)+str(j+1)+" "+tmp[0]]
                                if x - y > 0:
                                    y = x
                                    c = "20"+str(k)+str(j+1)
                            except:
                                pass
                    if c:
                        tmp.insert(0, c)
#                         print(tmp,df1[df1[0]==f[i]].iat[0, 1])
                        a = df1[df1[0] == f[i]].index[0]
                        df1.at[a, 1] = " ".join(tmp)

            elif len(tmp) == 3:
                if type(df2.at[df2[df2[0] == f[i]].index[0], 1]) == float:
                    if re.search("^20[0-9]+", tmp[0]):
                        y = df1_number_count_dic[tmp[0]+" "+tmp[1]]
                        x, c = 0, 0
                        for j in range(10):
                            for k in range(2):
                                try:
                                    x = df1_number_count_dic["20" +
                                                             str(k)+str(j)+" "+tmp[1]]
                                    if x - y > 0:
                                        y = x
                                        c = "20"+str(k)+str(j)
                                except:
                                    pass
                        if c:
                            tmp[0] = c
    #                         print(tmp, "/",df1[df1[0]==f[i]].iat[0, 1])
                            a = df1[df1[0] == f[i]].index[0]
                            df1.at[a, 1] = " ".join(tmp)
                try:
                    if df1_number_count_dic[tmp[1]+" "+tmp[2]] > 3:
                        if df1_number_count_dic["20"+data] > 3:
                            #                             print("20"+data)
                            a = df1[df1[0] == f[i]].index[0]
                            df1.at[a, 1] = "20"+data
                except:
                    pass
            elif len(tmp) == 1:
                if type(df2.at[df2[df2[0] == f[i]].index[0], 1]) == float:
                    x, y, c = 0, 0, 0
                    for j in range(10):
                        try:
                            x = df1_number_count_dic["0"+str(j)+" "+tmp[0]]
                            if x - y > 0:
                                y = x
                                c = "0"+str(j)
                        except:
                            pass
                    if c:
                        tmp.insert(0, c)
    #                     print(tmp, "/",df1[df1[0]==f[i]].iat[0, 1])
                        a = df1[df1[0] == f[i]].index[0]
                        df1.at[a, 1] = " ".join(tmp)
```
